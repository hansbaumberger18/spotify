# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/JOINT.csv')

print(df.shape)
df.head()
```

---
https://www.youtube.com/watch?v=8d7ywKCm6HI&list=TLPQMTgwMTIwMjLOQNHMg9wb3w&index=3

```python
# Check for null values
pd.isnull(df).sum()
```

```python
# Drop rows containing null values as they aren't a significant number
df.dropna(inplace = True)
df.shape
```

```python
df.info()
```

```python
df.drop_duplicates(['id'], inplace = True)
df.shape
```

```python
df.describe().transpose()
```

```python
# 10 least popular songs in the dataset
least_popular = df.sort_values('popularity', ascending = True)
least_popular.head(10)
```

```python
# 10 most popular songs in the dataset
most_popular = least_popular.tail(10)
most_popular
```

```python
#change release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'])

#If wanted release_date as index:
    #df.set_index('release_date', inplace = True)
```

```python
#Change duration in ms to seconds and drop duration_ms column
df['duration'] = df['duration_ms'].apply(lambda x: round(x/1000))
df.drop(columns = ['duration_ms'], inplace = True)
df.head(1)
```

```python
df.columns
```

```python
#Map the correlation of features
corr_df = df[['popularity', 'release_date',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature',
       'duration']].corr()

plt.figure(figsize = (15,8))
heatmap = sns.heatmap(corr_df,
                     annot = True,
                      vmin= -1,
                      vmax = 1,
                     cmap = 'inferno',
                     linecolor='Black')
heatmap.set_title('Correlation Heatmap between features');
```

```python
# Scatter subplots between different features
plt.figure(figsize = (12,10))

sns.regplot(data=corr_df, y = 'loudness', x = 'energy').set(title = 'Loudness vs Energy')
```

```python
# Number of songs each year
df['release_year'] = pd.DatetimeIndex(df['release_date']).year

sns.displot(df['release_year'], discrete = True, kind = 'hist').set(title = 'Number of songs per year');
```

```python
df.head(1)
```

```python
# Avg duration of songs during the years

song_duration = df.duration
sns.set_style(style = 'white')

fig, ax = plt.subplots(figsize = (18, 8))
fig = sns.barplot( x = df['release_date'], y = song_duration, ax=ax).set(title = 'Year vs Song duration')
#plt.xticks(rotation = 60);
```

```python
figures = plt.figure(figsize = (15,15))

#Danceability
ax1 = figures.add_subplot()
ax1.set_xlabel('Danceability')
ax1.set_ylabel('Count')
ax1.set_title('Song danceability distribution')
df.hist(column = ['danceability'], bins = 40);

#Release date
ax1 = figures.add_subplot()
ax1.set_xlabel('Release date')
ax1.set_ylabel('Count')
ax1.set_title('Song release dates')
df.hist(column = ['release_date'], bins = 40);

#Energy
ax1 = figures.add_subplot()
ax1.set_xlabel('Energy')
ax1.set_ylabel('Count')
ax1.set_title('Song energy')
df.hist(column = ['energy'], bins = 40);

#Key
ax1 = figures.add_subplot()
ax1.set_xlabel('Key')
ax1.set_ylabel('Count')
ax1.set_title('Song key')
df.hist(column = ['key'], bins = 40);

#Loudness
ax1 = figures.add_subplot()
ax1.set_xlabel('Loudness')
ax1.set_ylabel('Count')
ax1.set_title('Song loudness')
df.hist(column = ['loudness'], bins = 40);

#Mode
ax1 = figures.add_subplot()
ax1.set_xlabel('Mode')
ax1.set_ylabel('Count')
ax1.set_title('Song mode')
df.hist(column = ['mode'], bins = 40);

#Acousticness
ax1 = figures.add_subplot()
ax1.set_xlabel('Acousticnesss')
ax1.set_ylabel('Count')
ax1.set_title('Song acousticness')
df.hist(column = ['acousticness'], bins = 40);

#Instrumentalness
ax1 = figures.add_subplot()
ax1.set_xlabel('Instrumentalness')
ax1.set_ylabel('Count')
ax1.set_title('Song instrumentalness')
df.hist(column = ['instrumentalness'], bins = 40);

#Liveness
ax1 = figures.add_subplot()
ax1.set_xlabel('Liveness')
ax1.set_ylabel('Count')
ax1.set_title('Song liveness')
df.hist(column = ['liveness'], bins = 40);

#Valence
ax1 = figures.add_subplot()
ax1.set_xlabel('Valence')
ax1.set_ylabel('Count')
ax1.set_title('Song valence')
df.hist(column = ['valence'], bins = 40);

#Tempo
ax1 = figures.add_subplot()
ax1.set_xlabel('Tempo')
ax1.set_ylabel('Count')
ax1.set_title('Song tempo')
df.hist(column = ['tempo'], bins = 40);

#Time signature
ax1 = figures.add_subplot()
ax1.set_xlabel('Time signature')
ax1.set_ylabel('Count')
ax1.set_title('Song time signature')
df.hist(column = ['time_signature'], bins = 40);

#Duration
ax1 = figures.add_subplot()
ax1.set_xlabel('Duration')
ax1.set_ylabel('Count')
ax1.set_title('Song duration')
df.hist(column = ['duration'], bins = 40);

```
